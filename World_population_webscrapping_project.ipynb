```python

import bs4
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.worldometers.info/world-population/population-by-country/"
page=requests.get(url)
page.encoding = 'utf-8'    #useful for encoding sysmbols % - + etc
print(page)
```

```python
soup=BeautifulSoup(page.text,'html')
soup
```

```python
table=soup.find('table')
table
```

```python
rows=soup.find_all('tr')
rows
```

```python
population_count = []

for row in rows:   # skip header
    cols = row.find_all('td')
    pop=[p.text for p in cols]
    population_count.append(pop)
population_count
```

```python
population_count.pop(0)
```

```python
population_count
```

```python
Sr_no=[]
Country=[]
Population2025=[]
Yearly_change=[]
Net_change=[]
Density=[]
Land_area=[]
Migrants=[]
Fert_rate=[]
Median_age=[]
Urban_pop_percent=[]
World_share=[]
for i in population_count:
    Sr_no.append(i[0])
    Country.append(i[1])
    Population2025.append(i[2])
    Yearly_change.append(i[3])
    Net_change.append(i[4])
    Density.append(i[5])
    Land_area.append(i[6])
    Migrants.append(i[7])
    Fert_rate.append(i[8])
    Median_age.append(i[9])
    Urban_pop_percent.append(i[10])
    World_share.append(i[11])
Data={'Sr_no':Sr_no,'Country':Country,'Population2025':Population2025,'Yearly_change':Yearly_change,'Net_change':Net_change,'Density':Density,'Land_area':Land_area,'Migrants':Migrants,'Fert_rate':Fert_rate,'Median_age':Median_age,'Urban_pop':Urban_pop_percent,'World_share':World_share}
    
```

```python
Data
```

```python
import pandas as pd
import numpy as np
Population_Summary=pd.DataFrame(Data)
```

```python
Population_Summary
```

```python
Population_Summary.info()
```

```python
Population_Summary.isnull().sum()
```

```python
Population_Summary=Population_Summary.drop(['Sr_no'],axis=1)
```

```python
Population_Summary
```

```python
Population_Summary['Population2025']=Population_Summary['Population2025'].str.replace(',','')
Population_Summary['Population2025']=Population_Summary['Population2025'].astype('int')
Population_Summary
```

```python
Population_Summary['Yearly_change']=Population_Summary['Yearly_change'].str.replace('%','')
Population_Summary['Yearly_change']=Population_Summary['Yearly_change'].str.replace('−', '-')
Population_Summary['Yearly_change']=Population_Summary['Yearly_change'].str.strip()
```

```python
Population_Summary['Yearly_change']=Population_Summary['Yearly_change'].astype('float')
```

```python
Population_Summary
```

```python
Population_Summary['Net_change']=Population_Summary['Net_change'].str.replace(',','')
Population_Summary['Net_change']=Population_Summary['Net_change'].str.replace('%','')
Population_Summary['Net_change']=Population_Summary['Net_change'].str.replace('−','-')
Population_Summary['Net_change']=Population_Summary['Net_change'].astype(float)

Population_Summary.info()
```

```python
Population_Summary['Density']=Population_Summary['Density'].str.replace(',','')
Population_Summary['Density']=Population_Summary['Density'].astype(float)
Population_Summary.info()
```

```python
Population_Summary['Land_area']=Population_Summary['Land_area'].str.replace(',','')
Population_Summary['Land_area']=Population_Summary['Land_area'].astype(float)
Population_Summary.info()
```

```python
Population_Summary['Migrants']=Population_Summary['Migrants'].str.replace(',','')
Population_Summary['Migrants']=Population_Summary['Migrants'].str.replace('−','-')
Population_Summary['Migrants']=Population_Summary['Migrants'].astype(float)
Population_Summary.info()
```

```python
Population_Summary['Fert_rate']=Population_Summary['Fert_rate'].str.replace(',','')
Population_Summary['Fert_rate']=Population_Summary['Fert_rate'].astype(float)
Population_Summary.info()
```

```python
Population_Summary['Median_age']=Population_Summary['Median_age'].str.replace(',','')
Population_Summary['Median_age']=Population_Summary['Median_age'].astype(float)
Population_Summary.info()
```

```python
Population_Summary['Urban_pop']=Population_Summary['Urban_pop'].str.replace('%','')
Population_Summary['Urban_pop']=Population_Summary['Urban_pop'].str.strip()
Population_Summary['Urban_pop']=pd.to_numeric(Population_Summary['Urban_pop'],errors='coerce')  #astype(float) creashes on '' while to_numerice converts invalid to NaN
Population_Summary['Urban_pop']=Population_Summary['Urban_pop'].astype(float)

Population_Summary.info()
```

```python
Population_Summary['World_share']=Population_Summary['World_share'].str.replace('%','')
Population_Summary['World_share']=Population_Summary['World_share'].astype(float)
Population_Summary
```
